# Preparation Data

```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1FRoh0QXDp3BLS4gZX8MmhPWEm5j433Jf/export?format=csv"
df = pd.read_csv(url)

df.head()
```

```python
df.columns.tolist()
```

```python
df = df.drop(columns=['Email Address','Alamat Perusahaan','Nama Lengkap','Alamat KTP','Alamat Domisili','Divisi/Sub Divisi/Bagian Perusahaan','Nomor HP',
                      'Merged Doc ID - Hasil Skrining Prov Sumatera Barat','Merged Doc URL - Hasil Skrining','Link to merged Doc - Hasil Skrining Prov Sumatera Barat',
                      'Document Merge Status - Hasil Skrining','X'])
```

```python
df.columns.tolist()
```

```python
df['Tanggal Lahir'] = pd.to_datetime(
    df['Tanggal Lahir'],errors='coerce',dayfirst=True)
```

```python
df
```

```python
for i, column in enumerate(df.columns):
    print(f"Column: '{column}', Index: {i}")
```

```python
cols = df.columns[23:29].tolist() + [df.columns[41]]
df_logreg = df[cols]
df_logreg.head()
```

```python
df_logreg.columns.tolist()
```

```python
df_logreg['Hasil Kriteria'].value_counts()
```

```python
df_logreg['Hasil Kriteria'] = df_logreg['Hasil Kriteria'].replace({
    'Terduga 1': 'Terduga',
    'Terduga 2': 'Terduga',
    'Terduga 3': 'Terduga'
})
```

```python
df_logreg['Hasil Kriteria'].value_counts()
```

```python
tb_risk = ['Kekurangan Gizi',
 'Merokok',
 'Perokok Pasif',
 'Riwayat Diabetes Melitus/Kencing Manis',
 'Lansia Lebih dari 65 Tahun',
 'Ibu Hamil']
```

```python
X = df_logreg[tb_risk].replace({
    'Ya': 1,
    'Tidak': 0})
```

```python
X.head()
X.dtypes
```

```python
y = df_logreg['Hasil Kriteria'].map({
    'Terduga': 1,
    'Tidak Terduga': 0
})
```

```python
y.head()
y.dtypes
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y   # penting untuk data klasifikasi
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    solver='liblinear'
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    accuracy_score
)

y_pred = model.predict(X_test)

print("Accuracy :", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd

cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm,
    index=['Actual Tdk Terduga', 'Actual Terduga'],
    columns=['Pred Tdk Terduga', 'Pred Terduga']
)

cm_df
```

```python
coef_df = pd.DataFrame({
    'Faktor Risiko': X.columns,
    'Koefisien': model.coef_[0],
    'Odds Ratio': np.exp(model.coef_[0])
}).sort_values(by='Odds Ratio', ascending=False)

coef_df
```

```python
y_prob = model.predict_proba(X_test)[:, 1]
y_pred_03 = (y_prob >= 0.3).astype(int)

print(confusion_matrix(y_test, y_pred_03))
print(classification_report(y_test, y_pred_03))
```

```python
# visualisasi odds ratio

import matplotlib.pyplot as plt

coef_df = pd.DataFrame({
    'Faktor Risiko': X.columns,
    'Odds Ratio': np.exp(model.coef_[0])
}).sort_values(by='Odds Ratio', ascending=True)

plt.figure(figsize=(8,5))
plt.barh(coef_df['Faktor Risiko'], coef_df['Odds Ratio'])
plt.axvline(1, linestyle='--')  # garis netral
plt.xlabel('Odds Ratio')
plt.title('Odds Ratio Faktor Risiko TB')
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_prob = model.predict_proba(X_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.3f}')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('ROC Curve Logistic Regression TB')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import recall_score

thresholds = np.arange(0.1, 0.9, 0.05)

result = []
for t in thresholds:
    y_pred_t = (y_prob >= t).astype(int)
    recall = recall_score(y_test, y_pred_t)
    result.append((t, recall))

thr_df = pd.DataFrame(result, columns=['Threshold', 'Recall_Terduga'])
thr_df
```

```python
best_thr = thr_df.loc[thr_df['Recall_Terduga'].idxmax(), 'Threshold']
best_thr
```

```python
y_pred_best = (y_prob >= best_thr).astype(int)
confusion_matrix(y_test, y_pred_best)
```
